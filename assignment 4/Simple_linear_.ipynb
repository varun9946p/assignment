QUESTION 01

Predict delivery time using sorting time


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
```

```python
data= pd.read_csv('delivery_time.csv')
```

```python
data.head()
```

```python
delivery_time=data['Delivery Time']
Sorting_time=data['Sorting Time']
```

```python
data.info()
```

```python
data.isnull().sum
```

```python
model = smf.ols('delivery_time~Sorting_time',data=data).fit()
model.summary()
```

```python
model1 = smf.ols("delivery_time~np.log (Sorting_time)", data = data).fit()
model1.summary()
```

```python
model2 = smf.ols("np.log(delivery_time)~Sorting_time", data = data).fit()
model2.summary()
```

```python
model3 = smf.ols("np.log(delivery_time)~Sorting_time+pow(Sorting_time,2)+pow(Sorting_time,3)", data = data).fit()
model3.summary()
```

QUESTION 02

To build a prediction model of salary_hike

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('Salary_Data.csv')
```

```python
data.head()
```

```python
data.corr()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Salary'], data['YearsExperience'])
plt.xlabel('Salary')
plt.ylabel('YearsExperience')
plt.show()
```

```python
model4 = smf.ols('YearsExperience~Salary',data=data).fit()
model4.summary()
```

```python
plt.scatter(data['Salary'],model4.predict(data['Salary']),label='model 4 prediction', color='red')
```

```python
model4 = smf.ols("YearsExperience~Salary", data = data).fit()
model4.summary()
```

```python
data["Salary_sq"] = data.Salary**2
model4 = smf.ols("np.log(YearsExperience)~Salary+Salary_sq", data = data).fit()
model4.summary()
```
