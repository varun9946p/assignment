```python
Unit_A=data['Unit A']
Unit_B=data['Unit B']
```

```python
#TO FIND P VALUE AND T TEST
t_statistics,p_value = stats.ttest_ind(Unit_A,Unit_B)
print("t_statistics:",t_statistics)
print("p_value:",p_value)
```

The p value which we got here is less compared to the normal p value . So we reject the null hypothesis
