Here the given data is a catogarical one so we use chi-square test to examine the difference between the samples . The null and alternative hypothesis is already mentioned in the question.

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data=pd.read_csv('BuyerRatio (1).csv')
```

```python
data.info()
```

```python
data.head()
```

```python
observed_values = np.array([
    [50,142,131,70],
    [435,1523,1356,750]
])
```

```python
chi2_stat,p_value,df,expected_values = stats.chi2_contingency(observed_values)

print("Chi_squared statistic:",chi2_stat)
print("p-value:",p_value)
print("Degree of freedom:",df)
print("Expected_values:",expected_values)
```

here the p value is more than 0.5 therefore we will acceot the null hypothesis
