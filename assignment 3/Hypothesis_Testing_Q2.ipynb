From the question we take 'null hypothesis' as the reports of 4 leboratory is same and 'alternative hypothesis' as the report of 4 leboratory is not same
here we have more than 2 data so we take annova test

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data=pd.read_csv('LabTAT.csv')
```

```python
data.head()
```

```python
f_statistic, p_value = stats.f_oneway(data['Laboratory 1'], data['Laboratory 2'], data['Laboratory 3'], data['Laboratory 4'])
print("f_statistic:",f_statistic)
print("p_value",p_value)
```

here p value is more than 0.5 so we accept the null value. Therefore the reports of 4 laboratory is same
